```python
population_per_countries = pd.read_csv("data/countries.csv")
```

```python
uk_countries = population_per_countries[population_per_countries.country == "United Kingdom"]
```

```python
plt.plot(uk_countries.year, uk_countries.population / 10**6)
plt.xlabel("YEAR")
plt.ylabel("POPULATION (M)")
plt.title("Population variation of UK")
plt.show()
```
